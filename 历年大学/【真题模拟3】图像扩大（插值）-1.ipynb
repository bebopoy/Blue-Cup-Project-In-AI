# 手打答案
```python 
import numpy as np

def nearest_neighbor_interpolation(image, new_width, new_height):
    # old_width, old_height = image.shape 是错误的，先width 后height
    old_height, old_width = image.shape

    width_scale = new_width / old_width
    height_scale = new_height / old_height

    new_image = np.empty((new_height,new_width),dtype=np.uint8)

    for i in range(new_height):
        for j in range(new_width):
            raw_height_postion = int(i / height_scale)
            raw_width_postion = int(j / width_scale)
            
            new_image[i,j] = image[raw_height_postion,raw_width_postion]

    return new_image

def main():
    #5X5 的图像
    image = np.array([[56, 23, 15, 1, 3],
                      [65, 32, 78, 255, 0], 
                      [12, 45, 62, 1, 128],
                      [255, 0, 0,0,1],
                      [1, 128, 255,0,1]], dtype=np.uint8)
    #放大为 10X10
    resized_image = nearest_neighbor_interpolation(image, 10, 10)
    print(resized_image)

if __name__ == '__main__':
    main()
```
