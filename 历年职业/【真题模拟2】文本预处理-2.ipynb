# 文本预处理 word2Id  
word to index (数字)

文本数据预处理
介绍
随着社交媒体和数字内容的快速发展，新媒体平台成为人们获取信息、交流观点和表达意见的重要渠道。在新媒体平台上进行文本分析和情感分析的需求也越来越大，这些任务训练的模型可以帮助企业、组织或个人了解用户的反馈和偏好。为了更好地适应模型的需求，提高模型的性能和效果，对文本进行预处理是常见且必要的环节。

准备
word2id.txt ，是本任务提供的词汇与 id 相对应的集合，每行是一个词汇及其 id， 单词与 id 之间以符号 \t 相隔。
task.py ，是你后续答题过程中编写代码的地方。
目标
请按要求实现以下目标。 

目标 1：实现 read_word2id 函数

参数
file_path（字符串）：word2id.txt 文件路径。
功能
读取 word2id.txt 文件，并以字典形式返回。
返回值
word2id（字典）：字典中的 key 为单个词汇，value 为其对应的 id。形如：{'i': 0, '内容': 1,...}。
目标 2：实现 tokens_to_ids 函数

参数
tokens（列表）：输入的词汇列表，即一个字符串文本经分词处理后的形式。
word2id（字典）：词汇与其 id 一一对应的字典，其中 key 为单个词汇，value 为其对应的 id。
max_len（整数）：表示要输出的 id 列表的最大长度，默认为 10。
unk（字符串）：表示未登录词的标记，默认为 <unk> 。
pad（字符串）：表示填充符号的标记，默认为 <pad> 。
功能
将词汇列表转换为对应的 id 列表，设置固定长度 max_len。
如果列表长度小于 max_len，则用 <pad> 的 id 填充到列表末尾，使其长度为 max_len；
如果列表长度大于 max_len，则从列表的末尾开始截断，使其长度为 max_len。
返回值
ids（列表）：包含了将词汇转换为对应 id 后的结果。

输入示例：

['这个', '电影', '着实', '精彩']

输出示例：

[19, 9, 104, 60, 103, 103, 103, 103, 103, 103]

请在 task.py 文件中，基于以下代码，实现目标函数。

```python
import jieba

comments = [
    {"text": "这个电影真是好看了，剧情精彩，演员演技出众，特效惊人，yyds！", "language": "zh"},
    {"text": "真的很失望，剧情太拖沓了，演员演技平平，特效一般般。", "language": "zh"},
    {"text": "这本书写得棒，内容深入浅出，观点独到！", "language": "zh"},
    {"text": "这文章实在是糟糕，内容浅薄，观点站不住脚啊。", "language": "zh"},
    {"text": "I just watched the new movie and it blew my mind ! The plot twists were unexpected and kept me on the edge of my seat . The performances were outstanding , especially the lead actor . I would highly recommend it !", "language": "en"},
    {"text": "The food at this restaurant is absolutely delicious ! The flavors are well-balanced and the presentation is top-notch . The service is also excellent . I'll definitely be coming back !", "language": "en"},
    {"text": "I'm really impressed with this photo . The composition is stunning , and the lighting is perfect . It captures the beauty of the scenery beautifully . Great job , photographer !", "language": "en"}
]

word_lemma = {'watched': 'watch', 'blew': 'blow', 'twists': 'twist', 'were': 'be', 'kept': 'keep', 'me': 'i', 'performances': 'performance', 'is': 'be', 'flavors': 'flavor', 'are': 'be', 'coming': 'come', 'captures': 'capture'}

# 定义中文文本处理函数
def process_zh(text):
    seg_tokens = jieba.lcut(text)
    return seg_tokens

# 定义英文文本处理函数
def process_en(text):
    # 转换成小写
    text_lower = text.lower()
    # 按照" "进行分词
    text_tokens = text_lower.split(" ")
    lem_tokens = [word_lemma[i] if i in word_lemma else i for i in text_tokens]
    return lem_tokens

# 读取 word2id.txt 文件并转换为字典
def read_word2id(file_path):
    word2id = {}
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            line = line.strip()
            if line:
                word,idx = line.split('\t')
                word2id[word] = int(idx)
    return word2id

# 将分词后的列表转换为固定长度的 id 列表
def tokens_to_ids(tokens, word2id, max_len=10, unk='<unk>', pad='<pad>'):
    ids = []
    for token in tokens:
        if token in word2id:
            ids.append(word2id[token])
        else:
            # 如果单词不在 word2id 中，则用 <unk> 的 id 表示
            ids.append(word2id[unk])
    if len(ids) < max_len:
        # 如果列表长度小于 max_len，用 <pad> 的 id 进行填充
        ids += [word2id[pad]] * (max_len - len(ids))
    else:
        # 如果列表长度大于等于 max_len，则截断为 max_len 长度
        ids = ids[:max_len]
    return ids

# 获取输入文本的 id 列表
def get_input_ids(word2id_file_path):
    res = []
    word2id = read_word2id(word2id_file_path)
    for i in comments:
        tokens = []
        if i["language"] == "zh":
            tokens = process_zh(i["text"])  
        elif i["language"] == "en":
            tokens = process_en(i["text"])  
        else:
            print("error input language")
        ids = tokens_to_ids(tokens, word2id)
        res.append(ids)
    return res

def main():
    word2id_file_path = 'word2id.txt'
    res = get_input_ids(word2id_file_path)
    print(res)
if __name__ == '__main__':
    main()
```
